```python
import clip
import torch
from PIL import Image

# Load model + preprocess
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)

# Load and preprocess image
image_path = "./data/lund_1024/images/DJI_20241007102743_0369_V.JPG"
image = preprocess(Image.open(image_path)).unsqueeze(0).to(device)

# Define text prompts
text_prompts = ["a house", "a building", "a tree", "a road"]
text_tokens = clip.tokenize(text_prompts).to(device)

# Encode image and text
with torch.no_grad():
    image_features = model.encode_image(image)
    text_features = model.encode_text(text_tokens)

    # Normalize
    image_features /= image_features.norm(dim=-1, keepdim=True)
    text_features /= text_features.norm(dim=-1, keepdim=True)

    # Compute similarity
    similarity = (image_features @ text_features.T).squeeze(0)

# Print results
for prompt, score in zip(text_prompts, similarity):
    print(f"{prompt:15s}: {score.item():.3f}")
```

```python
import matplotlib.pyplot as plt

# Display the image and CLIP scores
plt.figure(figsize=(8, 6))
plt.imshow(Image.open(image_path))
plt.axis("off")
plt.title("CLIP Similarity Scores:\n" + "\n".join(
    [f"{p}: {s:.3f}" for p, s in zip(text_prompts, similarity.tolist())]),
    fontsize=12
)
plt.show()
```

```python
from transformers import CLIPSegProcessor, CLIPSegForImageSegmentation
from PIL import Image
import torch
import matplotlib.pyplot as plt

# Load CLIPSeg model
processor = CLIPSegProcessor.from_pretrained("CIDAS/clipseg-rd64-refined")
model = CLIPSegForImageSegmentation.from_pretrained("CIDAS/clipseg-rd64-refined")
model.eval()

# Load your image
image_path = "./data/lund_1024/images/DJI_20241007102743_0369_V.JPG"
image = Image.open(image_path).convert("RGB")

# Prompt
prompt = "a house"

# Prepare inputs
inputs = processor(text=prompt, images=image, return_tensors="pt")

# Run inference
with torch.no_grad():
    outputs = model(**inputs)
    preds = outputs.logits[0].sigmoid()  # Convert logits to probabilities

# Plot result
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(image)
plt.imshow(preds.numpy(), cmap="jet", alpha=0.5)
plt.axis("off")
plt.title(f"CLIPSeg Mask for: '{prompt}'")

plt.tight_layout()
plt.show()
```
